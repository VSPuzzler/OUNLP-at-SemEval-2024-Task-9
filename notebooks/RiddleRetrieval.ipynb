```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('all-MiniLM-L6-v2')
import openpyxl

wb = openpyxl.load_workbook('/content/EncodedQuestionsWP.xlsx')
ws = wb.active
for i in range(1,ws.max_row):
  embeddings = model.encode(ws["B"+str(i)].value)
  ws["C"+str(i)] = ','.join(map(str, embeddings))
  print(','.join(map(str, embeddings)))

wb.save('/output.xlsx')
```
