```python
import requests
def retrevial(URL):
  page = requests.get(URL)
  page = page.text
  extractedquestion = ""
  index = page.find('>Riddle:</strong>')
  index = index + 19
  flip= False
  while page[index] not in "?":
    if page[index] == "<" and flip == False:
      flip = True
    elif page[index] == ">" and flip == True:
      flip = False
    elif flip == True:
      flip = True
    else:
      extractedquestion = extractedquestion +page[index]
    index +=1
  extractedquestion = extractedquestion + "?"


  extractedanswer = ""
  index = page.find('>Answer<')
  while page[index] != ":":
    index +=1
  index = index + 1
  while page[index] not in "\<":
    extractedanswer = extractedanswer +page[index]
    index += 1

  return extractedanswer, extractedquestion
```

```python
import xlwt
from xlwt import Workbook
wb = Workbook()

# add_sheet is used to create sheet.
sheet1 = wb.add_sheet('Sheet 1')
sheet1.write(0, 0, "answer")
sheet1.write(0, 1, "question")

for i in range(1, 2):
  sheet1.write(i, 0, retrevial("https://www.riddles.com/"+str(i))[0])
  sheet1.write(i, 1, retrevial("https://www.riddles.com/10")[1])
  print(i)
wb.save('RiddleExtraction.xls')

```
